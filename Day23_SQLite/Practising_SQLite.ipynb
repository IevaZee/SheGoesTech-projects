```python
con = sqlite3.connect("chinook.db")
print(con)
```

```python
df = pd.read_sql_query("select * from tracks", con)
df.head()
```

```python
rock_df = pd.read_sql_query("select * from v_rock_songs", con)
rock_df.head()
```
